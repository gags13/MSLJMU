```python
base_path = '/content/drive/MyDrive/MS/dataset/'
data_path = base_path + 'final_filtered.csv'

sample_dataframe = pd.read_csv(data_path, index_col=False, delimiter=',')
sample_dataframe.drop(['Unnamed: 0'],axis=1, inplace=True)
sample_dataframe.head()
```

```python
sample_vm_ids = sample_dataframe.vmid.values
sample_vm_ids
```

```python
uber_df = pd.DataFrame()
```

```python
def createDF(checkVms, file):
  data_df = pd.read_csv(file, compression='gzip', names=["timestamp", "vm_id", "min_cpu", "max_cpu", "avg_cpu"])
  clean_list = [data_df]
  global uber_df
  uber_df = uber_df.append(data_df[data_df['vm_id'].isin(checkVms)])
  print(file + str(uber_df.shape))
  del clean_list
```

```python
prefix = '/content/drive/MyDrive/MS/dataset/cpu/vm_cpu_readings-file-'
postfix = '-of-195.csv.gz'
for i in range(1,196):
  s = prefix + str(i) + postfix
  createDF(sample_vm_ids, s)
```
